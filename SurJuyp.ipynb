```python

#import Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Now we need to read in the data
df = pd.read_csv(r'C:\Users\suresh\Desktop\PortfolioProject4\movies.csv')
```

```python
# Now let's take a look at the data

df
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Change data type
df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
df
```

```python
# Create correct year column
df['yearcorrect'] = df['released'].astype(str).str.split(',').str[-1]
df['yearcorrect'] = df['yearcorrect'].str.split(' ').str[1]
df
```

```python
# Sort by gross column
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', 20)
df
```

```python
# Remove duplicates
df.drop_duplicates()
```

```python
# Scatter plot between budget and gross 
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Budget for Films')
plt.ylabel('Gross Earning')
plt.show()
```

```python
# Plot using seaborn Earninig vs Gross
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Looking for correlation
df.corr() # here person method is default
#pearson, kendall, spearman
#df.corr(method='pearson')
#df.corr(method='kendall')
#df.corr(method='spearman')
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matric for numeric feature')
plt.xlabel('Movie feature')
plt.ylabel('Movie feature')
plt.show()
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matric for numeric feature')
plt.xlabel('Movie feature')
plt.ylabel('Movie feature')
plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
